### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Step 1: Extract and split database responses
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(';')
all_databases = [db.strip() for sublist in db_series for db in sublist]

# Step 2: Count frequencies
db_counts = pd.Series(all_databases).value_counts()

# Step 3: Select top 5
top_5 = db_counts.head(5)

# Step 4: Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Developers Want to Work With')
plt.axis('equal')  # Ensures it's a perfect circle
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Step 1: Extract and split developer types
dev_series = df['DevType'].dropna().str.split(';')
all_dev_roles = [role.strip() for sublist in dev_series for role in sublist]

# Step 2: Count role frequencies
dev_counts = pd.Series(all_dev_roles).value_counts()

# Step 3: Get top 5 roles
top_5_dev_roles = dev_counts.head(5)

# Step 4: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_dev_roles, labels=top_5_dev_roles.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Developer Roles in Survey')
plt.axis('equal')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# Use the correct column name with a space
os_series = df['OpSysProfessional use'].dropna().str.split(';')
all_os = [os.strip() for sublist in os_series for os in sublist]

# Count OS occurrences
os_counts = pd.Series(all_os).value_counts()

# Select top 5 operating systems
top_os = os_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
# Step 1: Clean and split the language entries
lang_series = df['LanguageHaveWorkedWith'].dropna().str.split(';')
all_languages = [lang.strip() for sublist in lang_series for lang in sublist]

# Step 2: Count language frequencies
lang_counts = pd.Series(all_languages).value_counts()

# Step 3: Select top 5 languages
top_langs = lang_counts.head(5)

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_langs, labels=top_langs.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Programming Languages Developers Have Worked With')
plt.axis('equal')  # Keeps the pie chart circular
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
# Step 1: Split and clean the collaboration tool entries
collab_series = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(';')
all_tools = [tool.strip() for sublist in collab_series for tool in sublist]

# Step 2: Count tool frequencies
tool_counts = pd.Series(all_tools).value_counts()

# Step 3: Grab the top 5 tools
top_tools = tool_counts.head(5)

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_tools, labels=top_tools.index, autopct='%1.1f%%', startangle=140)
plt.title('Top Collaboration Tools Used Professionally')
plt.axis('equal')
plt.show()
```
