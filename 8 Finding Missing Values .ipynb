### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Count missing values in each column
missing_counts = df.isnull().sum()

# Display columns with missing values (optional: filter out zeroes)
missing_counts = missing_counts[missing_counts > 0]

# Show the result
print("Missing values per column:")
print(missing_counts)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Calculate missing value percentages
missing_percent = df.isnull().mean() * 100

# Step 2: Filter columns with 10%–50% missing values
filtered_cols = missing_percent[(missing_percent > 10) & (missing_percent < 50)].index

# Step 3: Create heatmap for filtered columns
plt.figure(figsize=(14, 8))
sns.heatmap(df[filtered_cols].isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Missing Values Heatmap (10%-50% Missing)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size for better visibility
plt.figure(figsize=(12, 8))

# Create a heatmap of missing values
sns.heatmap(df.isnull(),
            cbar=False,
            cmap='YlGnBu',
            yticklabels=False)

plt.title("Missing Values Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_employment = df['Employment'].isnull().sum()
print(f"Missing rows in 'Employment': {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Find the most frequent value in the 'Employment' column
most_common_employment = df['Employment'].mode()[0]

print(f"The most frequent value in 'Employment' is: {most_common_employment}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Impute missing values in 'Employment' with the most frequent value
df['Employment'] = df['Employment'].fillna(most_common_employment)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
# Step 1: Count Employment categories
employment_counts = df['Employment'].value_counts()

# Step 2: Keep top 6, group rest as "Other"
top_n = 6
top_categories = employment_counts[:top_n]
other_total = employment_counts[top_n:].sum()

# Step 3: Create new Series with "Other"
cleaned_counts = top_categories.copy()
cleaned_counts["Other"] = other_total

# Horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(y=cleaned_counts.index, x=cleaned_counts.values, hue=cleaned_counts.index, palette='viridis')
plt.title("Employment Status Distribution (Horizontal)")
plt.xlabel("Count")
plt.ylabel("Employment Status")
plt.tight_layout()
plt.show()
```

```python
## Write your code here
# Count the frequency of each category in Employment
employment_counts = df['Employment'].value_counts()

# Plot a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=employment_counts.index, y=employment_counts.values, hue=employment_counts.index, palette='viridis', legend=False)
plt.title("Distribution of Employment Status After Imputation")
plt.xlabel("Employment Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
